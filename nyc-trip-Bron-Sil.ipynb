```python
df = spark.table("data_engineering.video1.yellow_tripdata_2015_01")
display(df.head(5))
```

```python
from pyspark.sql.functions import avg

df.select(avg("trip_distance")).show()
```

```python
df.orderBy(
    df.total_amount.desc()
).select(
    "trip_distance",
    "fare_amount",
    "total_amount"
).show(10)
```

```python
from pyspark.sql.functions import hour

hourly = (
    df.withColumn(
        "pickup_hour",
        hour("tpep_pickup_datetime")
    )
    .groupBy("pickup_hour")
    .count()
    .orderBy("pickup_hour")
)

display(hourly)
```

```python
clean_df = (
    df.filter(df.trip_distance > 0)
      .filter(df.fare_amount > 0)
      .filter(df.passenger_count > 0)
)
```

```python
print(df.count())
print(clean_df.count())
```

```python
from pyspark.sql.functions import hour

hourly_trips = (
    clean_df
    .withColumn("pickup_hour", hour("tpep_pickup_datetime"))
    .groupBy("pickup_hour")
    .count()
    .orderBy("pickup_hour")
)

display(hourly_trips)
```

```python
clean_df.write.mode("overwrite").saveAsTable(
    "data_engineering.video1.yellow_tripdata_silver"
)
```
