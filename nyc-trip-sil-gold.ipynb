```python
silver_df = spark.table(
    "data_engineering.video1.yellow_tripdata_silver"
)
```

```python
silver_df.show(5)
```

```python
from pyspark.sql.functions import month, sum

monthly_revenue = (
    silver_df
    .withColumn(
        "month",
        month("tpep_pickup_datetime")
    )
    .groupBy("month")
    .agg(
        sum("total_amount").alias("monthly_revenue")
    )
    .orderBy("month")
)

display(monthly_revenue)
```

```python
monthly_revenue.write.mode("overwrite").saveAsTable(
    "data_engineering.video1.gold_monthly_revenue"
)
```

```python
from pyspark.sql.functions import sum

payment_revenue = (
    silver_df
    .groupBy("payment_type")
    .agg(
        sum("total_amount").alias("revenue")
    )
)
```

```python
payment_revenue.write.mode("overwrite").saveAsTable(
    "data_engineering.video1.gold_payment_revenue"
)
```

```python
from pyspark.sql.functions import hour

hourly_trips = (
    silver_df
    .withColumn(
        "pickup_hour",
        hour("tpep_pickup_datetime")
    )
    .groupBy("pickup_hour")
    .count()
    .orderBy("pickup_hour")
)
```

```python
display(hourly_trips)
```

```python
hourly_trips.write.mode("overwrite").saveAsTable(
    "data_engineering.video1.gold_hourly_trips"
)
```
